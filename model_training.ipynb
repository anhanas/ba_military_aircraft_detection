```python
from ultralytics import YOLO
```

```python
model = YOLO("yolo26n-obb.pt")
```

```python
model.tune(
    data="data/data.yaml",
    iterations=5,
    epochs=10, 
    imgsz=960,
    batch=32,
    workers=4,
    patience=5
)
```

```python
model.nc = 20
results = model.train(
    data="data/data.yaml",
    epochs=100,
    imgsz=960,
    batch=32,
    patience=20,
    name="model_n-max",
    project="runs/train",
    task="obb",
    plots=True,
    save=True,
    device=0,
    cls_pw=1.0,
)
```
