```python
from ultralytics import YOLO
```

```python
# download pretrained model n
model_n = YOLO("yolo26n-obb.pt")
```

```python
model_n.nc = 20
```

```python
results = model_n.train(
    data="data/data.yaml",
    epochs=100,
    imgsz=640,
    batch=16,
    patience=20,
    name="model_n",
    project="runs/train",
    task="obb",
    plots=True,
    save=True,
    device=0,
    cls_pw=0.0,
)
```

```python
# after training: run validation/test and save plots
val_results = model_n.val(data="data/data.yaml", plots=True, save=True)
```

```python
model = YOLO("./runs/obb/runs/train/model_n-without-classweights/weights/best.pt")
model.val(
    data="data/data.yaml",
    plots=True,
    save=True,
    save_dir="./runs/obb/runs/val/model_n-without-classweights",
)
```

```python
preds = model.predict(
    source="./data/test/images",
    save=True,
    plots=True,
    save_dir="./runs/obb/runs/preds/model_n-without-classweights",
)
```

## Evaluating the Models on the Test-Split

```python
model_names = [
    "model_m-without-classweights",
    "model_m-with-classweights",
    "model_n-without-classweights",
    "model_n-with-classweights",
    "model_l-with-classweights",
    "model_x-with-classweights",
    "model_n-max",
]

for name in model_names:
    weights_path = f"./runs/obb/runs/train/{name}/weights/best.pt"
    model = YOLO(weights_path)
    metrics = model.val(
        data="data/data.yaml",
        split="test",
        plots=True,
        save=True,
        name=f"test_{name}",
        project="runs/val",
    )
```
